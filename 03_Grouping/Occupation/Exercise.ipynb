### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).

### Step 3. Assign it to a variable called users.

```python
url = r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep='|')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').agg({'age':'mean'}).reset_index()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
gb = users.groupby(['occupation','gender']).agg('count')['user_id'].reset_index()
gb.nunique()
```

```python
gb[gb['gender'] == 'M']['user_id'].reset_index(drop=True)
```

```python
count = gb[gb['gender'] == 'M']['user_id'].reset_index(drop=True) / gb[gb['gender'] == 'F']['user_id'].reset_index(drop=True).add(
    gb[gb['gender'] == 'M']['user_id'].reset_index(drop=True), fill_value=0)
count
```

```python
gbr = gb[gb['gender'] == 'M'].copy().reset_index()
gbr['ratio Male'] = count
gbr
```

```python
count_F = gb[gb['gender'] == 'F']['user_id'].reset_index(drop=True) / gb[gb['gender'] == 'F']['user_id'].reset_index(drop=True).add(
    gb[gb['gender'] == 'M']['user_id'].reset_index(drop=True), fill_value=0)
count.head()
```

```python
# Assuming your DataFrame is named df
# First, count the number of each gender for each occupation
counts = users.groupby(['occupation', 'gender']).size()
counts
```

```python
# Then, calculate the total count for each occupation
total = users.groupby('occupation').size()
total
```

```python
# Calculate the ratio for each gender within each occupation
rationing = counts / total
rationing = rationing.reset_index(name='ratio')
rationing
```

```python
usersRatio = pd.merge(users, rationing, on=['occupation', 'gender'], how='left')
usersRatio
```
