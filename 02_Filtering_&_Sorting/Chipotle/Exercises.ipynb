### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
lambd = lambda x:float(x[1:])

chipo['item_price'] = chipo['item_price'].apply(lambd)
```

```python
singleQuant = chipo.loc[chipo['quantity']==1,['item_name','item_price']]
singleQuant.head()
```

```python
maxItemPrice = singleQuant.groupby('item_name').agg({'item_price':'max'}).reset_index()
maxItemPrice
```

```python
(maxItemPrice['item_price']>10).sum()
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
maxItemPrice
```

### Step 6. Sort by the name of the item

```python
maxItemPrice.sort_values('item_name',ascending=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.loc[chipo['item_price'] == chipo['item_price'].max(),'quantity']
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
gbName = chipo.groupby('item_name').agg({'quantity':'sum'}).reset_index()
gbName[gbName['item_name'] == 'Veggie Salad Bowl']
```

```python
chipo[chipo.item_name == 'Veggie Salad Bowl'].quantity.sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
moreThanOne = chipo.loc[(chipo['item_name']=='Canned Soda') & (chipo['quantity']>1),['item_name','quantity']]
moreThanOne.count()
```
